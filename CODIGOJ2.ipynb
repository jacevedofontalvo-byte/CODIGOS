**CODIGO 1**

```python
# Autor: Estudiante - Semillero "Patrones en los Numeros Enteros"
# Descripcion:
# Programa que analiza los patrones en la suma de numeros pares e impares
# y permite observar posibles excepciones para formular una conjetura general.

def es_par(numero: int) -> bool:
    """Determina si un numero es par."""
    return numero % 2 == 0

def es_impar(numero: int) -> bool:
    """Determina si un numero es impar."""
    return numero % 2 != 0

def analizar_patrones(limite: int) -> None:
    print("\n=== ANALISIS DE SUMAS ENTRE NUMEROS PARES E IMPARES ===\n")

    excepcion_impares = False
    excepcion_pares = False
    excepcion_mixta = False

    # ---- Caso 1: Impar + Impar ----
    print("\n-> Caso 1: Suma de dos numeros impares\n")
    for a in range(1, limite + 1, 2):
        for b in range(1, limite + 1, 2):
            resultado = a + b
            print(f"{a} + {b} = {resultado}")
            # CORRECCIÓN: la excepción ocurre si NO da par
            if resultado % 2 != 0:
                excepcion_impares = True

    # ---- Caso 2: Par + Par ----
    print("\n-> Caso 2: Suma de dos numeros pares\n")
    for a in range(0, limite + 1, 2):
        for b in range(0, limite + 1, 2):
            resultado = a + b
            print(f"{a} + {b} = {resultado}")
            if resultado % 2 != 0:
                excepcion_pares = True

    # ---- Caso 3: Par + Impar ----
    print("\n-> Caso 3: Suma de un numero par y uno impar\n")
    for a in range(0, limite + 1, 2):
        for b in range(1, limite + 1, 2):
            resultado = a + b
            print(f"{a} + {b} = {resultado}")
            if resultado % 2 == 0:
                excepcion_mixta = True

    # ---- Conclusiones ----
    print("\n=== RESULTADOS DEL ANALISIS ===\n")

    if not excepcion_impares:
        print("✓ La suma de numeros impares siempre es par.")
    else:
        print("✗ Se encontraron excepciones en la suma de impares.")

    if not excepcion_pares:
        print("✓ La suma de numeros pares siempre es par.")
    else:
        print("✗ Se encontraron excepciones en la suma de pares.")

    if not excepcion_mixta:
        print("✓ La suma de un numero par y uno impar siempre es impar.")
    else:
        print("✗ Se encontraron excepciones en la suma mixta.")

# Llamada al programa
analizar_patrones(10)
```

```python
# Autor: Estudiante – Semillero "Patrones en los Números Enteros"
# Descripción:
# Programa visual que demuestra que la suma de los números impares consecutivos
# forma cuadrados perfectos. Usa círculos para representarlo gráficamente.

def mostrar_cuadro(n):
    """Dibuja un cuadrado con círculos de tamaño n x n."""
    for _ in range(n):
        print("⚪ " * n)
    print()  # Espacio entre cada representación


def suma_impares_cuadrados_visual(limite):
    """Muestra visual y numéricamente cómo los impares construyen cuadrados perfectos."""

    suma = 0           # Acumulador
    numero_impar = 1   # Primer número impar

    print("\n=== DEMOSTRACIÓN VISUAL DE CUADRADOS PERFECTOS ===\n")

    for i in range(1, limite + 1):
        suma += numero_impar
        cuadrado = i * i

        print(f"Paso {i}: suma de los {i} primeros impares = {suma}")
        print(f"Comparación: {suma} = {i}² ({cuadrado})\n")
        print(f"Representación visual del cuadrado {i}×{i}:\n")

        mostrar_cuadro(i)  # Dibujo del cuadrado

        numero_impar += 2  # Siguiente número impar

    print("Conclusión final:")
    print("✅ La suma de los n primeros números impares consecutivos es igual a n².")
    print("Cada número impar nuevo añade una fila y una columna al cuadrado anterior.")
    print("\nEjemplo: 1 + 3 + 5 + 7 + 9 = 25 = 5²\n")


# --- Punto de entrada ---
if __name__ == "__main__":
    print("\n--- PROGRAMA VISUAL: SUMA DE IMPARES Y CUADRADOS PERFECTOS ---\n")
    try:
        limite = int(input("¿Hasta qué número deseas comprobar (por ejemplo, 5)? "))
        if limite < 1:
            print("Por favor, ingrese un número entero positivo.")
        else:
            suma_impares_cuadrados_visual(limite)
    except ValueError:
        print("Error: debe ingresar un número entero válido.")
```
